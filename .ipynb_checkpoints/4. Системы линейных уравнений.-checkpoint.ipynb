### 1. Решить систему уравнений методом Крамера:

   **а)** $\begin{cases}
x_{1}-2x_{2}=1 \\
3x_{1}-4x_{2}=7
\end{cases}$

**Решение**


$detA\begin{vmatrix}
1 & -2 \\
3 & -4
\end{vmatrix} = (1\cdot-4) - (3\cdot-2) = -4 + 6 = 2$

$detA_1\begin{vmatrix}
1 & -2 \\
7 & -4
\end{vmatrix} = (1\cdot-4) - (7\cdot-2) = -4 + 14 = 10$

**$x_1=\frac{detA_1}{detA} \frac{10}{2} = 5$**

$detA_2\begin{vmatrix}
1 & 1 \\
3 & 7
\end{vmatrix} = (1\cdot7) - (3\cdot1) = 7 - 3 = 4$

**$x_2=\frac{detA_2}{detA} \frac{4}{2} = 2$**

```python
A = np.array([[1, -2], [3, -4]])
A1 = np.array([[1, -2], [7, -4]])
A2 = np.array([[1, 1], [3, 7]])
detA = np.round(np.linalg.det(A))
detA1 = np.round(np.linalg.det(A1))
detA2 = np.round(np.linalg.det(A2))
x1 = detA1/detA
x2 = detA2/detA
print(f'x1 = {x1}; x2 = {x2}')
```

__5*.__ Написать на Python программу с реализацией одного из изученных алгоритмов решения СЛАУ.

```python
def kramera(A, x):
    
    detA = np.round(np.linalg.det(A))
    
    dic = {}

    if np.size(A[0]) != np.sqrt(np.size(A)):
        print(f'Матрица не является квадратной.')
    else:
        for i in range(int(np.size(A[0]))):
            AT = copy.deepcopy(A.T)
            AT[i] = x
            det = np.round(np.linalg.det(AT.T))
            dic[f'x{i+1}'] = det/detA
    return dic
```
